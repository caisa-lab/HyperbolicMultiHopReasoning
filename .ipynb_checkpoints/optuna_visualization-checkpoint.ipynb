```python
import optuna
import matplotlib.pyplot as plt
sqlite_db_path = "sqlite:///optuna_study.db"
study_name = "optimize_curvature_lr"
study = optuna.load_study(study_name=study_name, storage=sqlite_db_path)

params=["c", "lr"]

fig = optuna.visualization.plot_contour(study, params=params)
fig.show()

fig = optuna.visualization.plot_param_importances(study)
fig.show()
fig = optuna.visualization.plot_rank(study, params=["lr", "c"])
fig.show()

fig = optuna.visualization.plot_slice(study, params=["lr", "c"])
fig.show()
```

```python
df = study.trials_dataframe()
import plotly.express as px
import numpy as np

# Assuming `df` is your DataFrame from Optuna trials
c_values = df['params_c']
lr_values = df['params_lr']
objective_values = df['value']

# Create the scatter plot
fig = px.scatter(
    x=lr_values, 
    y=c_values, 
    color=objective_values, 
    color_continuous_scale='Viridis', 
    labels={
        'x': 'Learning Rate (log scale)',
        'y': 'Objective Value',
        'color': 'Log of C values'
    },
    title='Custom Slice Plot with Log Scale'
)

# Set log scales
fig.update_xaxes(type='log')

# Show the plot
fig.show()
```
